Numerical calculator (if-else statements)

Design a simple numerical calculator in Python. The user provides threee inputs: two numbers and the mathmtical operator of their choice. The program should print the outputon the screen.



```python
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
op = input("Enter numerical operator: ")

if op == "+":
  res = num1 + num2
elif op == "-":
  res = num1 - num2
elif op == "*":
  res = num1 * num2
elif op == "/":
  res = num1 / num2
elif op == "//":
  res = num1 // num2
elif op == "**":
  res = num1 ** num2
elif op == "%":
  res = num1 % num2
else:
  print("Invalid operation")
  res = None

print(f"Result: {res}")
```

Parking Garage (for/while loop)


A parking garage charges $5 + $2.50 for each hour parked.

Write a program that prints the number of hours parked and the corresponding fee side-by-side, from 1 to 8 hours


Example output:

1. 7.5

2. 10.0

3. 12.5

4. 15.0

5. 17.5

6. 20.0

7. 22.5

8. 25.0

```python
base_fee = 5.0
hourly_rate = 2.50

print("Parking Fees:")
for hours in range(1, 9):
    fee = base_fee + (hourly_rate * hours)
    print(f"{hours}. {fee}")
```

Bonus: The minimum fee is 10 and the maximum fee is 20




```python
base_fee = 5.0
hourly_rate = 2.50
min_fee = 10.0
max_fee = 20.0

print("Parking Fees:")
for hours in range(1, 9):
    fee = base_fee + (hourly_rate * hours)
    if fee < min_fee:
        fee = min_fee
    elif fee > max_fee:
        fee = max_fee
    print(f"{hours}. {fee}")
```

Rock-Paper-Scissors (while loops)


```python
import random

play = input("Do you want to play a game? (yes/no): ").lower()

while play == "yes":
    p_move = input("Choose between rock, paper, or scissors: ").lower()
    c_move = random.choice(['rock', 'paper', 'scissors'])
    print("computer move: ", c_move)

    if p_move == c_move:
      print("Draw")
    elif (
        (p_move == "rock" and c_move == "scissors") or
        (p_move == "scissors" and c_move == "paper") or
        (p_move == "paper" and c_move == "rock")
        ):
      print("Player wins")
    else:
      print("Computer wins")

play = input("Do you want to play again? (yes/no): ")
```
