Numerical calculator (if-else statements)

Design a simple numerical calculator in Python. The user provides threee inputs: two numbers and the mathmtical operator of their choice. The program should print the outputon the screen.



```python
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
op = input("Enter numerical operator: ")

if op == "+":
  res = num1 + num2
elif op == "-":
  res = num1 - num2
elif op == "*":
  res = num1 * num2
elif op == "/":
  res = num1 / num2
elif op == "//":
  res = num1 // num2
elif op == "**":
  res = num1 ** num2
elif op == "%":
  res = num1 % num2
else:
  print("Invalid operation")
  res = None

print(f"Result: {res}")
```

Parking Garage (for/while loop)


A parking garage charges $5 + $2.50 for each hour parked.

Write a program that prints the number of hours parked and the corresponding fee side-by-side, from 1 to 8 hours


Example output:

1. 7.5

2. 10.0

3. 12.5

4. 15.0

5. 17.5

6. 20.0

7. 22.5

8. 25.0

```python
base_fee = 5.0
hourly_rate = 2.50

print("Parking Fees:")
for hours in range(1, 9):
    fee = base_fee + (hourly_rate * hours)
    print(f"{hours}. {fee}")
```

Bonus: The minimum fee is 10 and the maximum fee is 20



```python
base_fee = 5.0
hourly_rate = 2.50
min_fee = 10.0
max_fee = 20.0

print("Parking Fees:")
for hours in range(1, 9):
    fee = base_fee + (hourly_rate * hours)
    if fee < min_fee:
        fee = min_fee
    elif fee > max_fee:
        fee = max_fee
    print(f"{hours}. {fee}")
```

Rock-Paper-Scissors (while loops)


```python
import random

play = input("Do you want to play? (y/n): ").lower()

while play == "y":
    p_move = input("Choose between rock, paper, scissors: ").lower()
    c_move = random.choice(['rock', 'paper', 'scissors'])

    print(f"Player move: {p_move}")
    print(f"Computer move: {c_move}")

    if p_move == c_move:
        print("Draw")
    elif (
        (p_move == "rock" and c_move == "scissors") or
        (p_move == "scissors" and c_move == "paper") or
        (p_move == "paper" and c_move == "rock")
    ):
        print("Player wins!")
    else:
        print("Computer wins!")

    play = input("Do you want to play again? (y/n): ").lower()
```

```python
def roi(revenue, cost):
    assert cost != 0
    assert isinstance(revenue, (int, float))
    assert isinstance(cost, (int, float))
    return (revenue - cost) / cost
```

```python
def loan_payment(principal, annual_rate, months):
    assert isinstance(principal, (int, float))
    assert isinstance(annual_rate, (int, float))
    assert isinstance(months, int)

    if principal <= 0 or annual_rate <= 0 or months <= 0:
        raise ValueError("Invalid input values")

    try:
        r = annual_rate / 12 / 100
        if r == 0:
            pmt = principal / months
        else:
            pmt = r * principal / (1 - (1 + r) ** (-months))
    except Exception as e:
        raise Exception(f"Error: {e}")

    return pmt
```

```python
class Greeting:
    def __init__(self, first, last):
        self.first = first
        self.last = last

    def display(self, formal=False):
        if formal:
            print(f"{self.last}, {self.first}")
        else:
            print(f"{self.first}")

    def __str__(self):
        return "Class Greeting \nReturns formal or informal greetings"


s = Greeting("John", "Doe")
print(s)
s.display()
s.display(formal=True)
```

```python
class Student:
    def __init__(self, name, number):
        self.__name = name
        self.number = number
        self.__courses = []

    def get_courses(self):
        return self.__courses

    def get_name(self):
        return self.__name

    def set_name(self, newName):
        self.__name = newName

    def __len__(self):
        return len(self.__courses)

    def enroll(self, course):
        if course in self.__courses:
            print(f"Student {self.__name} has already enrolled in course {course}.")
        else:
            self.__courses.append(course)
```

```python
s = Student("Jane Doe", 12345)
print("Student name:", s.get_name())
s.set_name("Jane B. Doe")
print("Student name:", s.get_name())
print("Courses enrolled in:", s.get_courses())
s.enroll('BTE320')
s.enroll('BTE320')
print(f"Student {s.get_name()} has enrolled in {len(s)} course(s): {s.get_courses()}")
```
