Numerical calculator (if-else statements)

Design a simple numerical calculator in Python. The user provides threee inputs: two numbers and the mathmtical operator of their choice. The program should print the outputon the screen.



```python
num1 = float(input("Enter the first number: "))
operator = input("Enter an operator (+, -, *, /): ")
num2 = float(input("Enter the second number: "))

result = None

if operator == '+':
    result = num1 + num2
elif operator == '-':
    result = num1 - num2
elif operator == '*':
    result = num1 * num2
elif operator == '/':
    if num2 != 0:
        result = num1 / num2
    else:
        print("Error: Division by zero is not allowed.")
else:
    print("Error: Invalid operator.")

if result is not None:
    print(f"Result: {num1} {operator} {num2} = {result}")
```
